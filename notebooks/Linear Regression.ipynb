# linear regression model
### This notebook illustrate how to perform simple linear regression with the OLS model

#### Importing our libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

#### Load The Diamond Csv

```python
df=pd.read_csv('diamond.csv')
```

```python
df.head()
```

```python
#the dependent variable
Y=df['price']
x=df['carat']
```

```python
Y.head()
```

```python
X=sm.add_constant(x)
```

```python
print(X.head())
```

```python
model=sm.OLS(endog=Y,exog=X).fit()
```

```python
print(model.summary())
```
